```python
def split_text(text):
    return text.split(',')
sbert_model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')
topic_model = BERTopic.load('Jiranuwat/topic_model',embedding_model=sbert_model)
```

```python
data = pd.read_csv('articles_data.csv')
data
```

```python
hierarchical_topics = pd.read_csv('hierarchical_topics.csv')
hierarchical_topics
```

## cleaning

```python
from pprint import pprint
from tqdm import tqdm
import re
import string
import unicodedata
pd.options.mode.chained_assignment = None
translator = str.maketrans('', '', string.punctuation)
from pythainlp.corpus import thai_stopwords
from pythainlp.corpus.common import thai_words
import deepcut
from pythainlp.util import normalize
from pythainlp.corpus import thai_stopwords
from attacut import tokenize, Tokenizer
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
    # Replace English characters with an empty string
    #english_pattern = re.compile(r'[a-zA-Z]')
    #result = re.sub(english_pattern, '', text)
    result = re.sub(r'\d+', '', text)
    result = result.split(',')
    result = [item for item in result if item != '']

    thai__stopwords = list(thai_stopwords())+'EMagazine GJ international bangkok hospital'.lower().split(' ')
    result = [i for i in result if i not in thai__stopwords]

    eng_stop_words = set(stopwords.words('english'))
    result = [word for word in result if word.lower() not in eng_stop_words]

    output = ''
    for i in result:
        output = i+','+output

    result = output

    return result
```

```python
token =  data['text_token'].apply(remove_stopwords)
```

# tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(tokenizer=split_text)
X_tfidf = tfidf.fit_transform(token).toarray()
vocab = tfidf.vocabulary_
reverse_vocab = {v:k for k,v in vocab.items()}
feature_names = tfidf.get_feature_names_out()
df_tfidf = pd.DataFrame(X_tfidf,columns=feature_names)
```

```python
words_importance = []
for i in tqdm(range(len(df_tfidf))):
    words_importance.append(df_tfidf.iloc[i][df_tfidf.iloc[i].sort_values(ascending=False) >= 0.2].index.to_list())
```

```python
lst = []
for i in words_importance:
    txt = ''
    for j in i:
        txt = j +','+txt
    
    txt = txt[:-1]
    lst.append(txt)
```

```python
doc_info = topic_model.get_document_info(data['text_token'],data)
doc_info['words_importance'] = lst
doc_info
```

```python
parent_all_id = []
parent_all_name = []

for i in range(110):
    current_id = str(i)
    parent_id = []
    parent_name = []

    for j in range(1000):
        left_child = hierarchical_topics[hierarchical_topics['Child_Left_ID'].astype('string') == current_id]
        right_child = hierarchical_topics[hierarchical_topics['Child_Right_ID'].astype('string') == current_id]

        if not left_child.empty:
            parent_id.append(left_child.iloc[0]['Parent_ID'])
            parent_name.append(left_child.iloc[0]['Parent_Name'])
            current_id = str(left_child.iloc[0]['Parent_ID'])
        elif not right_child.empty:
            parent_id.append(right_child.iloc[0]['Parent_ID'])
            parent_name.append(right_child.iloc[0]['Parent_Name'])
            current_id = str(right_child.iloc[0]['Parent_ID'])
        else:
            break
    
    parent_all_id.append(parent_id)
    parent_all_name.append(parent_name)
```

```python
parent_map = pd.DataFrame({'topic_id':list(range(110)),'parent_all_id':parent_all_id,'parent_all_name':parent_all_name})
parent_map = parent_map.set_index('topic_id')
parent_map
```

```python
doc_info.join(parent_map,on='Topic')
```
